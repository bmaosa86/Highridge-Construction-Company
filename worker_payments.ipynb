```python
import random  # Import the random module to generate random values

# Create an empty list to hold workers
workers = []

# Use a for loop to create 400 workers
for i in range(1, 401):
    worker = {
        'id': i,  # Each worker has a unique ID (1 to 400)
        'name': f'Worker_{i}',  # Each worker gets a name like Worker_1, Worker_2, etc.
        'salary': random.randint(5000, 35000),  # Random salary between $5,000 and $35,000
        'gender': random.choice(['Male', 'Female'])  # Randomly assign either Male or Female
    }
    workers.append(worker)  # Add each worker to the list

def generate_payment_slips(worker_list):
    for worker in worker_list:
        try:
            # Conditional Statements for Employee Level
            if 10000 < worker['salary'] < 20000:
                employee_level = "A1"  # Assign level A1 if salary is between $10,000 and $20,000
            elif 7500 < worker['salary'] < 30000 and worker['gender'] == "Female":
                employee_level = "A5-F"  # Assign level A5-F if salary is between $7,500 and $30,000 and gender is Female
            else:
                employee_level = "Unassigned"  # If neither condition is met, level is Unassigned

            # Print the payment slip
            print(f"Payment Slip for {worker['name']}:")
            print(f"ID: {worker['id']}")
            print(f"Salary: ${worker['salary']}")
            print(f"Gender: {worker['gender']}")
            print(f"Employee Level: {employee_level}")
            print("-" * 40)

        except Exception as e:
            # Handle errors gracefully
            print(f"Error generating payment slip for {worker['name']}: {e}")

generate_payment_slips(workers)
```
